# PCA Supervised Learning

In this case, we added an additional feature to the dataset, the actual seed number for each team during that season. Prior to using supervised learning algorithms to predict match results, we performed PCA to the new dataset. First off, we observed that out of the 15 features in the data set, 70.2% of the variance could be captured by the first 3 principal components. Furthermore, 95% of the variance could only be captured by using at least 9 components
